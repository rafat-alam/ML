## 01. Introduction to Pandas

```python
import numpy as np
import pandas as pd
```

### Pandas Series

```python
data = np.array(['r', 'a', 'f', 'a', 't'])
s = pd.Series(data)
s
```

### Pandas DataFrame

```python
lst = ["rafat", "is", "a", "good", "boy"]
df = pd.DataFrame(lst)
df
```

### Load CSV File

```python
df = pd.read_csv("data.csv")
df.head()
```

### DF Info

```python
df.info()
```

### Count of Null Values

```python
df.isnull().sum()
```

### Handling Missing Data

```python
df = df.fillna(0)
df.head()
```

### Selecting and Filtering Data

```python
age = df[df['Age'] > 25]
age.head()
```

### Adding Columns

```python
age['A+W'] = age['Age'] + age['Weight']
age.head()
```

### Grouping Data (GroupBy)

```python
res = df.groupby('College')['Salary'].sum()
res
```

### isnull()

```python
dict = {
          'First Score' : [100, 90, np.nan, 95],
          'Second Score' : [30, 45, 56, np.nan],
          'Third Score' : [np.nan, 40, 80, 98]
       }

df = pd.DataFrame(dict)
df.isnull()
```

### notnull()

```python
df.notnull()
```

### Fill Missing Values

```python
temp = df
temp.fillna(0)
```

```python
temp = df
temp.replace(np.nan, 1)
```

### Drop Missing Values

```python
temp = df
temp.dropna()
```

### Iterating Over Rows
